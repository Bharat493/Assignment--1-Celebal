## 🏗️ Database Schema & Sample Data

````python
conn.executescript("""
DROP TABLE IF EXISTS order_items; DROP TABLE IF EXISTS orders;
DROP TABLE IF EXISTS products;    DROP TABLE IF EXISTS customers;

CREATE TABLE customers (
    customer_id INT PRIMARY KEY,
    first_name  VARCHAR(50)   NOT NULL,
    last_name   VARCHAR(50)   NOT NULL,
    email       VARCHAR(100)  UNIQUE NOT NULL,
    city        VARCHAR(50)   NOT NULL,
    state       VARCHAR(50)   NOT NULL,
    join_date   DATE          NOT NULL,
    is_premium  BOOLEAN       DEFAULT FALSE
);
CREATE TABLE products (
    product_id   INT           PRIMARY KEY,
    product_name VARCHAR(100)  NOT NULL,
    category     VARCHAR(50)   NOT NULL,
    brand        VARCHAR(50)   NOT NULL,
    unit_price   DECIMAL(10,2) NOT NULL CHECK (unit_price > 0),
    stock_qty    INT           NOT NULL DEFAULT 0 CHECK (stock_qty >= 0)
);
CREATE TABLE orders (
    order_id     INT           PRIMARY KEY,
    customer_id  INT           NOT NULL,
    order_date   DATE          NOT NULL,
    status       VARCHAR(20)   NOT NULL DEFAULT 'Pending'
        CHECK (status IN ('Pending','Shipped','Delivered','Cancelled')),
    total_amount DECIMAL(12,2) NOT NULL CHECK (total_amount >= 0),
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);
CREATE TABLE order_items (
    item_id      INT           PRIMARY KEY,
    order_id     INT           NOT NULL,
    product_id   INT           NOT NULL,
    quantity     INT           NOT NULL CHECK (quantity > 0),
    unit_price   DECIMAL(10,2) NOT NULL CHECK (unit_price > 0),
    discount_pct DECIMAL(5,2)  DEFAULT 0 CHECK (discount_pct BETWEEN 0 AND 100),
    FOREIGN KEY (order_id)   REFERENCES orders(order_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id)
);
""")
print("✅ All 4 tables created!")
````

### Q4. Primary Keys of each table & why they must be UNIQUE and NOT NULL

| Table | Primary Key |
|---|---|
| customers | customer_id |
| products | product_id |
| orders | order_id |
| order_items | item_id |

**Why UNIQUE:** Duplicate PKs would make rows indistinguishable. Two customers with `customer_id=101` means you can't determine whose order belongs to whom.

**Why NOT NULL:** A NULL PK row has no identity — it can't be referenced by foreign keys, precisely updated, or reliably deleted. The DB enforces NOT NULL on all PKs automatically.

### Q5. Constraints on email column — what happens on duplicate insert?

````python
try:
    conn.execute("""INSERT INTO customers VALUES
        (109,'Test','User','aarav.s@email.com','Delhi','Delhi','2024-09-01',0)""")
    conn.commit()
except Exception as e:
    print(f"❌ Error: {e}")
    print("→ UNIQUE constraint on email rejected the duplicate insert entirely.")
````

> 💡 **Insight:** The UNIQUE constraint on email ensures one account per email — just like any real e-commerce platform. The NOT NULL constraint ensures no customer can be created without an email.

### Q6. Insert a product with unit_price = -50 — what happens?

````python
try:
    conn.execute("""INSERT INTO products VALUES
        (209,'Broken Item','Electronics','NoName',-50.00,100)""")
    conn.commit()
except Exception as e:
    print(f"❌ Error: {e}")
    print("→ CHECK (unit_price > 0) constraint evaluated (-50 > 0) = FALSE → INSERT rejected.")
````

> 💡 **Insight:** CHECK constraints enforce business logic at the DB layer — negative prices are logically invalid. This is safer than relying only on application-level validation.

### Q11. Explain idx_orders_date — how does it speed up date-filter queries?

`CREATE INDEX idx_orders_date ON orders(order_date)` builds a **B-Tree** data structure sorted by date.

| | Without Index | With Index |
|---|---|---|
| Method | Full Table Scan | Index Range Seek |
| Complexity | O(n) | O(log n + k) |
| 1M rows | Slow | Fast |

**Query that directly benefits:**

````python
run_query("""
SELECT order_id, customer_id, status, total_amount
FROM orders
WHERE order_date BETWEEN '2024-08-10' AND '2024-08-25'
""")
# idx_orders_date lets the engine seek directly to Aug 10
# and scan forward to Aug 25 — no full table scan
````

### Q12. Would YEAR(join_date)=2024 use the index? Rewrite as SARGable.

**❌ Non-SARGable (index NOT used):**
```sql
SELECT * FROM customers WHERE YEAR(join_date) = 2024;
```
Wrapping the column in YEAR() means the index (built on raw dates) cannot be used → full table scan.

**✅ SARGable rewrite (index IS used):**

````python
run_query("""
SELECT * FROM customers
WHERE join_date >= '2024-01-01'
  AND join_date < '2025-01-01'
""")
# Column is unwrapped — optimizer can use index for range seek
````

> 💡 **Rule:** Never apply a function to an indexed column in WHERE. Transform the constant on the right instead. SARGable = **S**earch **ARG**ument **ABLE**.

## 📕 Section D — Joins & Relationships

### Q19. INNER JOIN — orders with customer names.

````python
run_query("""
SELECT o.order_id, o.order_date, c.first_name, c.last_name, o.total_amount
FROM orders o
INNER JOIN customers c ON o.customer_id = c.customer_id
""")
````

> 💡 **Insight:** All 10 orders matched. Aarav Sharma and Rohan Gupta each placed 2 orders — repeat customers worth loyalty program targeting. INNER JOIN returns only rows with matches in both tables.

### Q20. LEFT JOIN — ALL customers with orders (NULL if no orders).

````python
run_query("""
SELECT c.customer_id, c.first_name, c.last_name,
       o.order_id, o.order_date, o.status, o.total_amount
FROM customers c
LEFT JOIN orders o ON c.customer_id = o.customer_id
""")
````

> 💡 **Insight:** All 8 customers appear even if they had no orders. In this dataset everyone has ordered. LEFT JOIN guarantees every left-table row is preserved — critical for finding inactive customers.

### Q21. Three-table JOIN: orders → order_items → products.

````python
run_query("""
SELECT o.order_id, p.product_name, oi.quantity, oi.unit_price, oi.discount_pct
FROM orders o
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p     ON oi.product_id = p.product_id
ORDER BY o.order_id
""")
````

> 💡 **Insight:** 15 line items across 10 orders. Running Shoes (₹4599) and Smart Watch (₹2999) are the highest-value items. Some items have discounts (5–15%) applied.

### Q22. LEFT JOIN vs RIGHT JOIN vs FULL OUTER JOIN

| Join | Returns | When to Use |
|---|---|---|
| **LEFT JOIN** | All left rows + matching right (NULL if no match) | Show all customers, even those with no orders |
| **RIGHT JOIN** | All right rows + matching left (NULL if no match) | Show all orders, even orphaned ones |
| **FULL OUTER JOIN** | All rows from both tables (NULL on either side) | Audit — find unmatched records on both sides simultaneously |

```sql
-- FULL OUTER JOIN in MySQL (emulated with UNION):
SELECT c.first_name, o.order_id
FROM customers c LEFT JOIN orders o ON c.customer_id = o.customer_id
UNION
SELECT c.first_name, o.order_id
FROM customers c RIGHT JOIN orders o ON c.customer_id = o.customer_id;
```
> ⚠️ MySQL doesn't natively support FULL OUTER JOIN — use UNION of LEFT + RIGHT JOINs.

| Child Table | FK Column | Parent Table | Referenced PK |
|---|---|---|---|
| orders | customer_id | customers | customer_id |
| order_items | order_id | orders | order_id |

## 📓 Section E — Advanced Concepts

### Q24. CASE statement — price tier classification.

````python
run_query("""
SELECT product_name, unit_price,
  CASE
    WHEN unit_price < 1000                THEN 'Budget'
    WHEN unit_price BETWEEN 1000 AND 3000 THEN 'Mid-Range'
    ELSE                                       'Premium'
  END AS price_tier
FROM products ORDER BY unit_price
""")
````

> 💡 **Insight:** 3 Budget products, 3 Mid-Range, 2 Premium. CASE works like if-elif-else inside SQL — evaluated top to bottom, first match wins.

### Q25. CASE inside aggregate — Delivered vs Not Delivered in one row.

````python
run_query("""
SELECT
  SUM(CASE WHEN status = 'Delivered'  THEN 1 ELSE 0 END) AS delivered,
  SUM(CASE WHEN status != 'Delivered' THEN 1 ELSE 0 END) AS not_delivered
FROM orders
""")
````

> 💡 **Insight:** 6 delivered, 4 not (2 Shipped + 1 Pending + 1 Cancelled). This conditional aggregation computes both counts in a single query pass — more efficient than running two separate COUNT queries.

### Q26. ACID Properties with Real-World Examples

| Property | Meaning | Bank Transfer Example |
|---|---|---|
| **A — Atomicity** | All-or-nothing. Every operation in the transaction succeeds, or none apply. | Debit ₹5000 from A AND Credit ₹5000 to B must both succeed. If credit fails → debit auto-rolls back. Money can't vanish. |
| **C — Consistency** | DB moves from one valid state to another. All constraints satisfied before & after. | Total money in the system stays the same. CHECK (balance >= 0) prevents negative balances. |
| **I — Isolation** | Concurrent transactions don't see each other's intermediate states. | Two tellers transferring from Account A simultaneously — locks prevent double-spending. |
| **D — Durability** | Once committed, data survives crashes. Written to persistent storage. | After "Transfer successful", a power cut won't undo it — it's in the transaction log on disk. |

> 💡 **ShopEase context:** When a customer places an order, ACID ensures: the order record + stock deduction + payment charge all happen atomically (A), inventory never goes negative (C), two customers can't buy the last item simultaneously (I), and confirmed orders survive server restarts (D).

### Q27. Complete Transaction — insert order 1011 atomically with ROLLBACK on failure.

````python
try:
    conn.execute("BEGIN TRANSACTION")
    
    # Step 1: Insert the new order
    conn.execute("""INSERT INTO orders (order_id, customer_id, order_date, status, total_amount)
        VALUES (1011, 102, '2026-05-27', 'Pending', 1598.00)""")
    print("✅ Step 1: Order 1011 inserted")
    
    # Step 2a: First order item — Wireless Earbuds x1
    conn.execute("""INSERT INTO order_items (item_id, order_id, product_id, quantity, unit_price, discount_pct)
        VALUES (5016, 1011, 201, 1, 1499.00, 0)""")
    print("✅ Step 2a: Wireless Earbuds item inserted")
    
    # Step 2b: Second order item — Cushion Covers x1
    conn.execute("""INSERT INTO order_items (item_id, order_id, product_id, quantity, unit_price, discount_pct)
        VALUES (5017, 1011, 208, 1, 599.00, 0)""")
    print("✅ Step 2b: Cushion Covers item inserted")
    
    # Step 3a: Reduce Wireless Earbuds stock by 1
    conn.execute("UPDATE products SET stock_qty = stock_qty - 1 WHERE product_id = 201")
    print("✅ Step 3a: Wireless Earbuds stock updated (250 → 249)")
    
    # Step 3b: Reduce Cushion Covers stock by 1
    conn.execute("UPDATE products SET stock_qty = stock_qty - 1 WHERE product_id = 208")
    print("✅ Step 3b: Cushion Covers stock updated (400 → 399)")
    
    # Step 4: All steps passed — COMMIT
    conn.execute("COMMIT")
    print("\n🎉 TRANSACTION COMMITTED SUCCESSFULLY!")

except Exception as e:
    conn.execute("ROLLBACK")
    print(f"\n❌ TRANSACTION FAILED: {e}")
    print("🔄 All changes rolled back — database unchanged.")
````

**Verify the transaction results:**

````python
print("New order created:")
display(run_query("SELECT * FROM orders WHERE order_id = 1011"))
print("\nUpdated stock:")
display(run_query("SELECT product_id, product_name, stock_qty FROM products WHERE product_id IN (201, 208)"))
````

> 💡 **Insight:** The ROLLBACK in except ensures atomicity — if ANY of the 5 operations fails (e.g., a duplicate item_id, stock going negative), ALL changes undo instantly. Either all 5 succeed and COMMIT, or nothing changes. This is ACID Atomicity in action.
