7. Save cleaned dataset + print summary report

## Step 3: Handle Missing Values

```python
print(' MISSING VALUES PER COLUMN')
missing = df.isnull().sum()
pct = (missing / len(df) * 100).round(1)
report = pd.DataFrame({'Missing Count': missing, 'Missing %': pct})
print(report[report['Missing Count'] > 0])
print(f'\nTotal missing values: {missing.sum()}')
```

```python
df = df.assign(discount=df['discount'].fillna(df['discount'].median()))
print(f'discount: filled 121 missing values with median = {df["discount"].median()}%')

df = df.assign(seller_name=df['seller_name'].fillna('Unknown Seller'))
print('seller_name: filled 301 missing values with "Unknown Seller"')

df = df.assign(what_customers_said=df['what_customers_said'].fillna('No Reviews'))
print('what_customers_said: filled 573 missing values with "No Reviews"')
```

```python
cols_to_drop = ['videos', 'seller_information', 'variations']
df.drop(columns=cols_to_drop, inplace=True)
print(f'Dropped columns: {cols_to_drop}')

before = len(df)
df.dropna(inplace=True)
print(f'\nRows before dropna : {before}')
print(f'Rows after dropna  : {len(df)}')
print(f'Missing remaining  : {df.isnull().sum().sum()}')
```

## Step 4: Basic Operations — Filter Rows & Select Columns

```python
df['final_price'] = df['final_price'].str.replace(r'[₹",\s]', '', regex=True).astype(float)
print('final_price cleaned. Sample:')
print(df[['title','initial_price','discount','final_price']].head(5))
```

```python
top_rated = df[df['rating'] >= 4.0]
print(f'Well-rated products (rating >= 4.0): {len(top_rated)}')
print(top_rated[['title','category','rating','final_price']].head(6))
```

```python
high_discount = df[df['discount'] >= 60]
print(f'Products with 60%+ discount: {len(high_discount)}')
print(high_discount[['title','category','initial_price','discount','final_price']].head(6))
```

```python
backpacks = df[df['category'] == 'backpacks']
print(f'Backpacks: {len(backpacks)} products')
print(backpacks[['title','rating','initial_price','final_price']].head(5))
```

```python
key_cols = ['product_id','title','product_description','category',
            'rating','ratings_count','initial_price','discount','final_price','seller_name']
selected = df[key_cols]
print('Selected Key Columns')
selected.head()
```

## Step 5: Remove Duplicates

```python
print(f'Total rows        : {len(df)}')
print(f'Duplicate rows    : {df.duplicated().sum()}')
print(f'Duplicate product_ids: {df["product_id"].duplicated().sum()}')
```

```python
df.drop_duplicates(inplace=True)
print(f'Rows after removing duplicates:{len(df)}')
```

## Step 6: Create Derived Column — total_amount

```python
df['total_amount'] = (df['initial_price'] * df['discount'] / 100).round(2)

print('total_amount column added (amount saved in ₹)')
df[['title','initial_price','discount','final_price','total_amount']].head(8)
```

```python
print('Discount Saved (total_amount) Statistics')
print(f'Min  savings : ₹{df["total_amount"].min():,.2f}')
print(f'Max  savings : ₹{df["total_amount"].max():,.2f}')
print(f'Mean savings : ₹{df["total_amount"].mean():,.2f}')
print(f'\n Average savings by category (top 5)')
print(df.groupby('category')['total_amount'].mean().sort_values(ascending=False).head(5))
```

## Step 7: Save Cleaned Dataset + Summary Report

```python
clean_cols = ['product_id','title','product_description','category',
              'rating','ratings_count','initial_price','discount',
              'final_price','total_amount','seller_name','currency']

df[clean_cols].to_csv('Combined_dataset_cleaned.csv', index=False)
print('Saved: Combined_dataset_cleaned.csv')
print(f'Final shape:{df[clean_cols].shape}')
print()
df[clean_cols].head()
```
